```python
# macOS系统python连接mysql
# 参考：https://github.com/PyMySQL/PyMySQL
import pymysql.cursors

# 连接数据库
connection = pymysql.connect(host='localhost',
                             user='root',
                             password='123456',
                             db='bigdata',
                             charset='utf8mb4',
                             cursorclass=pymysql.cursors.DictCursor)

try:
    with connection.cursor() as cursor:
        # 读取数据
        sql = "SELECT `location`, `counts` FROM `location_info`"
        cursor.execute(sql)
        result = cursor.fetchall()
finally:
    connection.close()
```

```python
# 处理数据，使其满足可视化的格式
value = []
attr = []
for i in range(len(result)):
    value.append(result[i]['counts'])
    attr.append(result[i]['location'].replace("省", "").replace("市", ""))
# 求出最大值，用于显示访问次数visual_range的max参数
maxValue = max(value)
```

```python
# 数据可视化
# 参考：https://github.com/pyecharts/pyecharts
#      http://pyecharts.org/#/zh-cn/prepare

from pyecharts import Map

map = Map("cdn访问ip统计", width=800, height=600)
map.add("", attr, value, maptype='china', is_visualmap=True,
        visual_text_color='#000',visual_range=[0, 8000])
map.render(path="/Users/gao/github/learningSpark/python/map.html")
map
```
